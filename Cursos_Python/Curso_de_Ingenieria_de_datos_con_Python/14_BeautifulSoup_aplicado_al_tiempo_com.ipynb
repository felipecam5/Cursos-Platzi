## Web Scrapping a eltiempo.com

```python
import requests
import bs4

url = "https://www.eltiempo.com"

response = requests.get(url)

soup = bs4.BeautifulSoup(response.text, 'html.parser')

# print(response.text)

news = soup.select('.title-container a')

print(news)

for item in news:
    print(f"{url}{item['href']}")
```

## Web Scrapping a github personal


```python
import requests
import bs4

response = requests.get("https://felipecam5.github.io/Cursos-Platzi/Curso_Practico_de_HTML_y_CSS/3_Clone_Google_Version_FCH/GOOGLE_CLONE_OFICIAL/index.html")
webpage = response.text

html = bs4.BeautifulSoup(response.text, "html.parser")

href_text1 = html.select(".footer-right a")
href_text2 = html.select(".footer-left a") 

print(href_text2)
links = [categoria.text for categoria in href_text1]

for categoria in href_text2:
    links.append(categoria.text) 
    
print(links)
```
