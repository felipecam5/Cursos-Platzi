# Extracción de información del HTML

```python
import requests
import bs4

response = requests.get("https://felipecam5.github.io/Cursos-Platzi/Curso_Practico_de_HTML_y_CSS/3_Clone_Google_Version_FCH/GOOGLE_CLONE_OFICIAL/index.html")

soup = bs4.BeautifulSoup(response.text, "html.parser")

print(soup)
# print(soup.title.text)
# print(soup.select("[class=menu-icon]"), soup.select("body"))
#print(soup.select("section[class=main-logo]") [0] ["class"])
# print(soup.select("section[class=main-logo]"))

#Imprimir todas las etiquetas href que tenga el documento:

paginas_links = soup.select(".footer-right") [0].text
paginas_links.strip

links = [link for link in paginas_links]


print(paginas_links, links)
```
